```python
#Importing Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset from an Excel file
df = pd.read_excel('data.xlsx')
```

```python
#Univariate Analysis

# Plot histogram for Salary
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True, bins=30)
plt.title('Salary Distribution')
plt.xlabel('Salary (INR)')
plt.ylabel('Frequency')
plt.show()

# Observation:
# The salary distribution appears to be right-skewed, which is common for salary data.
# A large number of salaries fall in the lower range (less than 300,000 INR), with very few high salaries.
#  There may be some extreme outliers with unusually high salaries.
```

```python
# Boxplot for detecting outliers in Salary
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Salary'])
plt.title('Boxplot of Salary')
plt.show()

# Observation:
# The boxplot shows several extreme outliers in salary data.
# Most salaries are concentrated in the lower range, while a few candidates earn much higher salaries.
# The presence of outliers suggests that we may need to handle them carefully during further analysis.
```

```python
# Check for outliers in AMCAT scores
amcat_scores = ['English', 'Logical', 'Quant', 'ComputerProgramming', 'Domain']
plt.figure(figsize=(10, 6))
df[amcat_scores].boxplot()
plt.title('Boxplot of AMCAT Scores')
plt.show()

# Observation:
# AMCAT scores show reasonable ranges across all skill sets (English, Logical, Quant, etc.).
# The boxplot reveals some outliers, especially in the "Domain" and "ComputerProgramming" sections.
# Most of the scores seem to be well-distributed, with no extreme variations.
```

```python
# Countplot for Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()

# Observation:
# The dataset shows an imbalance in gender representation, with more male candidates than female candidates.
# This may influence the results when analyzing gender-related salary outcomes or job roles.
```

```python
# Countplot for Specialization
plt.figure(figsize=(12, 6))
sns.countplot(y='Specialization', data=df, order=df['Specialization'].value_counts().index)
plt.title('Specialization Distribution')
plt.show()

# Observation:
# The most common specializations in the dataset are Computer Science Engineering, Electronics Engineering, and Information Technology.
# Some specializations (such as Textile Engineering, Metallurgical Engineering) have significantly fewer candidates.
# This imbalance should be noted during analysis, as some specializations may have too few candidates for statistically valid conclusions.
```

```python
# Countplot for Designation
plt.figure(figsize=(12, 6))
sns.countplot(y='Designation', data=df, order=df['Designation'].value_counts().index)
plt.title('Designation Distribution')
plt.show()
```

```python
#Bivariate Analysis

# Scatter plot of Salary vs collegeGPA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='collegeGPA', y='Salary', data=df)
plt.title('Scatter Plot of collegeGPA vs Salary')
plt.show()

# Observation:
# There is no clear linear relationship between College GPA and Salary.
# Some candidates with lower GPAs earn relatively high salaries, indicating that GPA may not be the sole determinant of salary.
# Further analysis may be required to identify other contributing factors such as job role or skillset.



# Pairplot to visualize relationships between multiple numerical variables
sns.pairplot(df[['Salary', 'collegeGPA', '10percentage', '12percentage', 'English', 'Logical', 'Quant']])
plt.show()

# Observation:
# There are no strong linear relationships between the different variables.
# Salary does not appear to have a clear correlation with academic scores (10th percentage, 12th percentage, GPA) or AMCAT scores (English, Logical, Quant).
# The scatter plots show a lot of variance, indicating multiple factors likely influence salary outcomes.
```

```python
# Boxplot of Salary vs Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('Boxplot of Salary by Gender')
plt.show()

# Observation:
# Male candidates appear to have a wider range of salaries compared to female candidates.
# The median salary for female candidates seems to be slightly lower than that of male candidates, though more analysis is needed to confirm statistical significance.
# Both genders have some extreme outliers in terms of high salaries.
```

```python
# Stacked bar plot for Gender vs Specialization
gender_specialization = pd.crosstab(df['Gender'], df['Specialization'])
gender_specialization.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Stacked Bar Plot: Gender vs Specialization')
plt.show()

# Observation:
# Male candidates are more prevalent in most specializations, particularly in core engineering fields like Mechanical, Electrical, and Civil Engineering.
# Female candidates are more concentrated in Computer Science and Information Technology specializations.
# This suggests potential gender-based preferences or trends in specialization selection.

# Another way to view categorical relationships using heatmaps
plt.figure(figsize=(12, 6))
sns.heatmap(pd.crosstab(df['Gender'], df['Specialization']), cmap='Blues', annot=True, fmt='d')
plt.title('Heatmap of Gender vs Specialization')
plt.show()
```

```python
# Boxplot of Salary vs Specialization
plt.figure(figsize=(14, 8))
sns.boxplot(x='Specialization', y='Salary', data=df)
plt.title('Boxplot of Salary by Specialization')
plt.xticks(rotation=90)
plt.show()

# Observation:
# Candidates with specializations in Computer Science and Information Technology have a higher median salary than other fields.
# Some fields like Electronics and Mechanical Engineering also show a broad range of salaries, with a few high-paying jobs.
# Specializations like Textile and Chemical Engineering have much lower salary ranges.
```

```python
# Boxplot of Salary vs Designation
plt.figure(figsize=(14, 8))
sns.boxplot(x='Designation', y='Salary', data=df)
plt.title('Boxplot of Salary by Designation')
plt.xticks(rotation=90)
plt.show()

# Observation:
# Software Engineers and Associate Engineers have the highest median salaries.
# Lower-paying roles include Technical Support Engineer and Data Analyst.
# There is a wide range of salaries for Software Engineers, indicating that this role may vary significantly depending on factors like experience or company.
```
